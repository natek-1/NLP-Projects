## Bag of words Inuitition

1. Create a list of all the words in the document
2. Convert each document(questions) into a vector containing the counts of each word

Limitations

1. Can be too many words in the dataset
2. Some words may occur to frequently
3. Some might occur to rarely or once(technical terms) can't allow us to make a lable 
4. A single word may have many forms(go, gone, going)

## Stop word removal
Removing communly occuring words

```python
english_stopwords = stopwords.words('english')
", ".join(english_stopwords)
```

```python
def remove_stopwords(tokens):
  return [word for word in tokens if not word.lower() in english_stopwords]
```

```python
q0_stp = remove_stopwords(q0_tok)
```

## configure the Count vectorizer parameter


```python
stemmer = SnowballStemmer(language='english')
```

```python
def tokenizer(text):
  return [stemmer.stem(word) for word in word_tokenize(text)]
```

```python
tokenizer('sexual sex biggest big boy boyfriend man manly real really')
```

```python
vectorizer = CountVectorizer(lowercase=True,
                             tokenizer=tokenizer,
                             stop_words=english_stopwords,
                             max_features=1000)
```

```python
%%time
vectorizer.fit(sub_df.question_text)
```

```python
vectorizer.get_feature_names_out()
```

```python
%%time
input = vectorizer.transform(sub_df.question_text)
```

## Train logistic regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
MAX_ITER = 1000
```

```python
model = LogisticRegression(max_iter=MAX_ITER, solver='sag')
```

```python
%%time
model.fit(train_inputs, train_target)
```

```python
train_pred = model.predict(train_inputs)
pd.Series(train_pred).value_counts(normalize=True)
```

```python
train_target.value_counts(normalize=True)
```
